4. In one year, I eat 27 apples. Write code to calculate how many apples I eat in 40 years, and round the answer to the closest hundred apples.

```python
apples = 27
apples_for_40_years = round(apples * 40, -2)
print(apples_for_40_years)
```

7. In one line of code, remove the question mark from the end of this sentence and change the sentence to capitalize only the first letter of the sentence:

```python
sentence = "THIS IS THE LAST QUESTION?"
new_sentence = sentence.rstrip("?").capitalize()
print(new_sentence)
```
